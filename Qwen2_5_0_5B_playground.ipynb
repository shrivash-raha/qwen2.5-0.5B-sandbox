# Looped generation

```python
import torch
from transformers import AutoTokenizer
from transformers import AutoModelForCausalLM

tokenizer= AutoTokenizer.from_pretrained("./qwen")
model = AutoModelForCausalLM.from_pretrained(
    "./qwen",
    torch_dtype="auto"
)
```

```python
n_top_tokens = 1
max_iters = 50
input_text = "Hello"
# input_text = "and here this sentence ends. Thank"
tokenized_input = tokenizer(input_text, return_tensors="pt")
print(tokenized_input)
temperature = 0.7

i = 0
while True:
  print(f'\n------{i+1}--------')
  outputs = model(**tokenized_input)

  scaled_logits = outputs.logits[0, -1] / temperature
  probs = torch.softmax(scaled_logits, dim = -1)

  # next_token_prob, next_token_id = torch.topk(probs, k=n_top_tokens, dim=-1)

  next_token_id = torch.multinomial(probs, num_samples=1)
  next_token_prob = probs[next_token_id]

  next_token = tokenizer.decode(next_token_id.item())

  print(f"{input_text}{next_token}[Token: \"{next_token}\", ID: {next_token_id.item()}, Prob: {next_token_prob.item()}]")

  input_text += next_token

  tokenized_input = tokenizer(input_text, return_tensors="pt")

  if next_token_id.item() == 151645:
    print(f"\n------EOS token {next_token_id.item()} encountered. Stopping generation.--------")
    break

  if i >= max_iters:
    print(f"\n------{max_iters} iterations reached------")
    break
  i += 1
```

```python
print(model.config.eos_token_id)
print(tokenizer.eos_token_id)
```

### Using `model.generate()` for more robust generation

To ensure proper stopping when an EOS token is expected, it's generally recommended to use the model's built-in `generate` method. This method handles token sampling, stopping conditions (like `eos_token_id` and `max_new_tokens`), and other generation parameters more efficiently.

```python
input_text_generate = "and here this sentence ends. Thank"
tokenized_input_generate = tokenizer(input_text_generate, return_tensors="pt")

# Get the EOS token ID from the tokenizer or config
eos_token_id = tokenizer.eos_token_id if tokenizer.eos_token_id is not None else config["eos_token_id"]

# Generate text using the model's generate method
output_tokens = model.generate(
    **tokenized_input_generate,
    max_new_tokens=200, # Max tokens to generate
    num_beams=1, # For greedy decoding
    do_sample=True, # To enable sampling
    top_k=50, # Sample from top 50 tokens
    temperature=0.7, # Adjust creativity
    eos_token_id=eos_token_id # Stop generation when EOS token is encountered
)

# Decode the generated tokens
generated_text = tokenizer.decode(output_tokens[0], skip_special_tokens=True)

print(generated_text)
```
